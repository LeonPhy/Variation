```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from astropy.io import ascii
from glob import glob
import tensorflow as tf
import h5py
```

```python
def constraint(x, y):
    return x**2+y**2-1

def fun(x, l, r, s):
    y = (np.sin(x[0]) + np.cos(x[1]+s))
    co = constraint(x[0],x[1])
    con = l*(co)
    penalty = 0.5*r*co**2
    f = y+con+penalty
    return f
```

```python
convergence_limit = 200
rhos = [0.01,0.1,0.5,1,10]
sym = 0.00

for rho in rhos:
    f, x, l = [], [np.array([-0.5,-0.5])], [1]
    #c = constraint(x[i][0],x[i][1])

    for i in range(convergence_limit):
        arg = {}
        arg['l'] = l[i]
        arg['r'] = rho
        arg['s'] = sym
        args = tuple(arg.values())
        print(args)
        res = minimize(fun,x[i],args=args)#,options={'disp':True})
        #if not res.success:
        #    print('Couldn\'t minimize the function in the {} iteration'.format(i+1))
        x.append(res.x)
        c = constraint(x[i+1][0],x[i+1][1])
        l.append(l[i]+rho*c)
        fu = fun(x[i+1], l[i+1], rho, sym)
        f.append(fu)
        sanity = np.dot((x[i]-x[i+1]),x[i]-x[i+1])**0.5 + abs(l[i+1] - l [i])
        #sanity = 1
        if sanity > 0.000001:
            continue
        else:
            print('Stopped at iteration {}'.format(i+1))
            break

    if i == convergence_limit-1:
        print("Didn't have enough iterations to converge!")
        
    [x0, x1] = x[-1]
    print(f'rho = {rho} : x = ({x[i][0]},{x[i][1]}), constraint = {x[i][0]**2+x[i][1]**2}, l = {l[i]}') 

    plt.figure(figsize=(12,8),facecolor='white')
    x_t = np.linspace(-2,2,100)
    y_t = np.meshgrid(x_t,x_t)
    theta = np.linspace(0,2*np.pi,100)
    plt.imshow(fun(y_t,0,rho,sym),extent=(-2,2,-2,2))
    ran = np.linspace(0,i+1,i+2)
    sol = np.array(x)
    plt.plot(sol[:,0],sol[:,1], color='red', marker='x')
    plt.show()
```
